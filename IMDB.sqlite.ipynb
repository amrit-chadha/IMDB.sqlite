```python
#Most voted movie in IMDB Database.

import sqlite3
import pandas as pd

db = sqlite3.connect('../imdb.sqlite')

data = pd.read_sql_query("Select Title, Rating,TotalVotes from IMDB",db)
max_votes = data['TotalVotes'].idxmax()
print(data.Title[max_votes],data.Rating[max_votes])
```

```python
#Movie with the Longest-Running Time from IMDB Movie database.

import sqlite3
import pandas as pd

db = sqlite3.connect('../imdb.sqlite')

data = pd.read_sql_query("Select Title, Runtime from IMDB",db)

data.fillna(0,inplace=True)
data['Runtime']= data['Runtime'].str.replace('min','')
data['Runtime'] = data['Runtime'].str.strip()
data.Runtime = pd.to_numeric(data.Runtime, errors='coerce')

max_rt = data['Runtime'].idxmax()
print(data.Title[max_rt],data.Runtime[max_rt])
```

```python
#Movie which has the maximum net profit in IMDB Movie Database.

import sqlite3
import pandas as pd

db = sqlite3.connect('../imdb.sqlite')

data = pd.read_sql_query("Select Title,Domestic,Worldwide,Budget from earning join IMDB on IMDB.Movie_id=earning.Movie_id",db)
data.fillna(0,inplace=True)

data['Profit'] = (data.Domestic+data.Worldwide) - pd.to_numeric(data.Budget,errors ='coerce')
data['Profit'].fillna(0,inplace=True)

max_pr = data['Profit'].idxmax()
print(data.Title[max_pr])
```

```python
#The least budgeted movie in IMDB Movie Database. 
#If there is more than one least budgeted movies then we print the movie which has maximum net profit.

import sqlite3
import pandas as pd

db = sqlite3.connect('../imdb.sqlite')

data = pd.read_sql_query("Select Title, Budget from IMDB",db)
data['Budget'] = pd.to_numeric(data.Budget,errors ='coerce')
lb = data['Budget'].idxmin()
print(data.Title[lb])
```
